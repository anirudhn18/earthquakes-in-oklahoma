```python
# Program to analysze earthquake data
# 
# Author - Shishir Shirbhate

from pandas import Series, DataFrame
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

# import plotly.plotly as py
# import plotly.tools as tls

%matplotlib inline

path = "D:/MS Material/3. Sem-1/5. Programming and Application Development - Reily Ken/Python Project/Data/"
# path_processed = "D:/MS Material/3. Sem-1/5. Programming and Application Development - Reily Ken/Python Project/Data/"

eq_data = pd.read_csv(path + 'earthquake_22454.csv',squeeze = True)

eq_data['origintime_new'] = pd.to_datetime(eq_data['origintime_new'])
eq_data['YEAR'] = eq_data['origintime_new'].apply(lambda x : x.year)
```

```python
# Graph of year and number of earthquakes according to years

eq_freq = eq_data[['YEAR','max_mmi']]

eq_freq = eq_freq[eq_freq.max_mmi !=0]
eq_freq = eq_freq[eq_freq.YEAR != 2016]
eq_freq = eq_freq[eq_freq.YEAR > 1970]

eq_freq = eq_freq.groupby(['YEAR']).count()

eq_freq.plot(legend=None)
plt.ylabel("No of earthquakes")
plt.show()
```

```python
path = "D:/MS Material/3. Sem-1/5. Programming and Application Development - Reily Ken/Python Project/Data/"

eq_data_other = pd.read_csv(path + 'earthquake_data.txt',squeeze = True)
```

```python
eq_other_freq = eq_data_other[['DateTime','Magnitude']]

eq_other_freq = eq_other_freq.replace('NaN',0)

eq_other_freq = eq_other_freq[eq_other_freq.Magnitude != 0]

eq_other_freq['DateTime'] = pd.to_datetime(eq_other_freq['DateTime'])

eq_other_freq['Year'] = pd.DatetimeIndex(eq_other_freq['DateTime']).year

eq_other_freq = eq_other_freq.groupby(['Year']).count()

eq_other_freq = eq_other_freq[eq_other_freq.index != 2016]

plt.xlabel("Years")
plt.ylabel("No of Earthquakes")

plt.plot(eq_other_freq.index,eq_other_freq['Magnitude'])
```

```python
# No of earthquakes in different county's from 2004-2015

eq_freq = eq_data[['YEAR','county','max_mmi']]

eq_freq = eq_freq[eq_freq.max_mmi !=0]
eq_freq = eq_freq[eq_freq.YEAR != 2016]
eq_freq = eq_freq[eq_freq.YEAR > 2004]

eq_freq = eq_freq.groupby(['county']).count()

eq_freq = eq_freq.sort('max_mmi')

eq_freq.plot(kind='bar',legend=None)
plt.ylabel("No of earthquakes")
plt.show()
```
